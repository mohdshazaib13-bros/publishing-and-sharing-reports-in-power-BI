2.  Define Power BI Service and explain its role in comparison to Power BI Desktop in the reporting lifecycle.

    - The reporting lifecycle typically includes data preparation → report creation → publishing → sharing → monitoring. Both tools play different but complementary roles: 1 - Data Preparation & Report Creation.
                             2 - Publishing Reports.
                             3 - Sharing & Collaboration.
                             4 - Data Refresh & Maintenance.
                             5 - Access & Consumption.
                             6 - Security & Governance.
   - Power BI Desktop is used to build and design reports.
     Power BI Service is used to publish, share, and manage those reports.

7.  Describe the four workspace roles available in Power BI and explain how access control is managed using these roles.

    - In the Power BI Service, workspaces use four roles to control what users can see and do. These roles help manage access control effectively.

     - Four Workspace Roles in Power BI:
     1- Admin
     - Full control over the workspace, Can add/remove users, assign roles, publish and delete content.
     2- Member
     - Can create, edit, and publish content, Has almost all permissions except full admin control.
     3- Contributor
     - Can create and modify content but has limited sharing control.
     4- Viewer
     - Read-only access to content.
     - Admins control who gets access and what level of permissions they have.Members and Contributors are given rights to create and manage content depending on their role in the project.Viewers are restricted to only consuming content, ensuring data security.

8.  Why are Power BI Apps considered more secure and scalable than direct sharing?Explain this from a governance and enterprise usage perspective.

    - Power BI Apps in the Power BI Service are considered more secure and scalable than direct report sharing because they are designed for centralized distribution, governance, and enterprise-wide control.
    - 1. Governance Perspective
     Power BI Apps: Content is managed in a single workspace and published as a controlled package (app). Only authorized creators can update it.
     Direct Sharing: Reports are shared individually, making it harder to track who has access.
    - 2. Enterprise Usage Perspective
     Apps: Designed to distribute content to large audiences (hundreds/thousands of users) efficiently.
     Direct Sharing: Requires adding users one-by-one or in small groups.


9.  Explain Row-Level Security (RLS) in Power BI.
How does RLS ensure that different users see different data using the same report?

    - Row-Level Security (RLS) in the Power BI Service is a feature that restricts data access at the row level based on the user viewing the report. It ensures that users only see the data they are authorized to see, even when using the same report.
    - How RLS Works:-
     1- Define Roles in Power BI Desktop.
     2- Publish the Report.
     3- Assign Users to Roles.
     4- Dynamic Data Filtering.
    - How RLS Ensures Different Users See Different Data
      The same dataset and report are used for all users.
      RLS applies user-specific filters behind the scenes.
      Each user sees only the rows that match their assigned role.
